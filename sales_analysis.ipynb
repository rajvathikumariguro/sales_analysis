```python
a=sns.countplot(x="Gender",data=ds)
for bars  in a.containers:
    a.bar_label(bars)
```

most buyers are females 


```python
a=sns.countplot(x="Age Group",data=ds)
for bars  in a.containers:
    a.bar_label(bars)
```

most buyers are from Age Group of 26-35


```python
a=sns.countplot(x="Age Group",data=ds,hue="Gender")
for bars  in a.containers:
    a.bar_label(bars)
```

most buyers are females from Age Group of 26-35


```python
state=ds.groupby(['Gender'],as_index=False)['Amount'].sum()
sns.barplot(data=state,x='Gender',y='Amount')
```

```python
state=ds.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=False).head(5)
sns.barplot(data=state,x='State',y='Amount')

sns.set(rc={'figure.figsize':(15,5)})
```

```python
state=ds.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by="Orders",ascending=False).head(10)
sns.barplot(data=state,x='State',y='Orders')
sns.set(rc={'figure.figsize':(15,5)})
```

The highest orders are from the state Uttar Pradesh
